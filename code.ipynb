```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(open("html_snippet.html", encoding='utf-8'), "html.parser")
```

```python
from utils import *

simplified_soup = simplize_html(soup)
```

```python
frequencies = calculate_frequency(simplified_soup)
```

```python
import pandas as pd

df = pd.DataFrame(list(frequencies.items()), columns=['Class', 'Frequency'])
df = df.sort_values('Frequency', ascending=False)
```

```python
df
```

```python
leave_classes = df['Class'].tolist().copy()
for cls in df['Class'].tolist():
    if cls is None:
        continue
    
    for leave_cls in leave_classes:
        if leave_cls is None:
            continue
        if cls == leave_cls:
            continue
        if cls in leave_cls and cls in leave_classes:
            leave_classes.remove(cls)
leave_classes
```
